#### Moving Data around

Size(Matrix) - returns dimenions of Matrix as Vector (Size(Matrix, 1) for first dim etc.)
Length(List) - for lists
who(s) - returns variables in current scope/workspace (s for more details)
clear variable - to clear variable
save filename variable - saves variable in file with filename (.mat, .txt -ascii etc.)

a = load Filename (be in the directory or add the path) to load it into a variable
oder
a = load('Filename')

A(3,2) - return matrix element row 3 column 2
A(3,:) - returns everything in row 3
A([1 3], :) - returns everything in row 1 and 3
-> Those can also be used for assignments
A = [A, [1 3; 1 4; 1 5]] -> to append a matrix
C = [A B] - concatenating matrices (or [A; B] for on top of each other)

#### Computational Operations on Data

A * B normal matrix multiplication
A .* B element-wise multiplication
A (.)^ 2 hoch zwei
1 ./ A 1/element (pro element)
exp(variable)
log(variable)
abs(variable) - absolute values
v + 1
v + ones(x, y)
A' - Transponierte Matrix
max(A) - returns max
[val, ind] = max(A) - returns value and index
min(A) - returns min
a<3 - element-wise check of this operation, returns 0 or 1 
find(expression) - find elements that fulfill the criteria
sum(a) sums all elements
floor(a) - round element up
ceil(a) - round element up
eye(dim) - gives the identity matrix
flipud - flips matrix up down
pinv(A) - returns inverse of A
